# 1. Data & Basic Libraries

Let's import basic libraries and the data we have in Section 1.4: **Algorithms and How They Work**.

```python
# Basic libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/5x12/ml4everyone/master/supplements/data/dataset_example1.csv', delimiter=',', header=0)
df.head(2)
```

```python
plt.scatter(df['x'], df['y'])
plt.show()
```
