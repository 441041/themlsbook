```python
import seaborn as sns
```

```python
popcorn_love = [34.3, 36.3, 37.2, 38.8, 12.2]
popcorn_not = [2.9, 3.1, 3.2, 4.2, 5.1]
```

```python
sns.distplot(popcorn_love)
sns.distplot(popcorn_not)
```

```python
p1 = sns.kdeplot(popcorn_love, bw=10, color="r")
p1 = sns.kdeplot(popcorn_not, bw=2, color="b")
```
